# Analyzing the Current Conversation Surrounding Virtual Reality and Its Intersection with Cybersecurity

In recent years, cybersecurity has become an increasing concern for governments, businesses, and individuals as technological advancements provide greater avenues for data breaches and the chance for financial gain, competitive advantages, or leverage provide compelling incentive for cybercrime. Issues like online trust and safety, mis- and dis-information, political interference, trade secret confidentiality, exposure of personally identifiable information, financial fraud, and economic and industrial espionage constitute major topics of discussion in the realm of cybersecurity. As artificial intelligence rapidly sophisticates, users globally are breaking new ground on discussing policies for usage, the recognition and negative implications of deep fakes, and the integrity of information on the internet.

A few years ago I attended a talk held by Accenture on the topic of the “Metaverse.” Having had no former exposure to this field, I found the explanation and imagination for the future of virtual reality groundbreaking. With useful applications spanning every industry – from medical to architectural to linguistic – virtual reality has the potential to change how humans interact in the world. However, it is not without its complexities and disadvantages as well. Consequently, as my graduate studies at Columbia have focused around the field of cybersecurity, I found myself curious to see how the conversation around virtual reality and its intersection with cybersecurity has changed over time.

Virtual reality has been slowly increasing in popularity and as a topic of conversation as it becomes steadily more refined and usable. Formerly, VR was presumably relegated to the world of gaming but it now has more tangible, daily applications. As VR improves and becomes increasingly integrated into reality, such as through Google Maps VR street-view overlay, policymakers should be concerned about the cyber implications. Cyber conflict is the new horizon of global threats and any technology that contributes to that space should be on policymakers’ radars.


### **THE HYPOTHESIS**

My initial hypothesis was that research into VR would reveal fewer cybersecurity conversations in the past but increasingly greater intersection in the present once technological advancement is refined enough that the social, geopolitical, and cyber risks become highly possible threats. To delve into this analysis, I originally intended to compare top news articles from 2014, 2019, and 2024 on the topics of virtual reality and cybersecurity. However, I quickly ran into error codes in Python, discovering that many publicly available news APIs do not allow extended historical access. Consequently, I decided to focus my research on the current news media conversation surrounding virtual reality; what are the major concerns, developments, and topics headlining the field right now? Fortuitously, this concentrated focus on the present conversation on VR revealed several notable insights that ultimately support my hypothesis that cybersecurity concerns only really enter public discussion after there is enough technological advancement in the field.

### **THE PROCESS**

Using the New York Times API  and NewsAPI.org, I conducted an initial search for the top 20 articles on “Virtual Reality AND Cybersecurity” in 2024 and in the past month, respectively. Intriguingly, from the NYT API, this led to a singular result about a current TikTok legal battle; I then broadened the search term to intersect “Virtual Reality AND Security” and discovered a whopping total of 4 relevant articles. NewsAPI.org generated 0 results. In short, these results indicated that there is not much conversation bonding these topics.

Therefore, to appropriately analyze the current conversation on virtual reality and understand why cybersecurity was not readily discussed in its context, I proceeded to set my parameters to search for “Virtual Reality” and “Cybersecurity” separately and narrowed the scope to articles under the technology section of the sites, as needed. As the NewsAPI allowed for a maximum of 100 articles within the past month, I initially tested my parameters with a large number of articles. This resulted in too many articles that were primarily advertisements infiltrating the data set. A bit of trial and error enabled me to land on 20 articles as ideal for analyzing the conversation while staying in scope of the project focus. To keep sample sizes the same, I utilized this same parameter in the NYT API. These parameters allowed me to pull the top 20 articles for “Virtual Reality” and “Cybersecurity,” respectively, visualize the most frequent terms in WordClouds and TF-IDF heatmaps, run topic modeling, and ultimately compare results.


*WORD CLOUDS*

To begin, I first visualized the major topics of discussion through the use of word clouds. With the New York Times API, I set the parameters to query the top 20 articles on “Virtual Reality” from January 1, 2024, to December 31, 2024. As the NYT API only allows 10 articles to be accessed at once, it was necessary to create a function that would access the first 10 articles then loop back around and add on the next 10, allowing for a final sample size of 20 articles. ChatGPT was useful in working out this code. After that, it was necessary to text_combine the article titles then generate the initial word cloud. In order to clean the data for the best results, I utilized the natural language processing via the natural language toolkit (NLTK), filtered out stopwords, lemmatized the articles, then repopulated the final word cloud. This same process was repeated with a query of “Cybersecurity” on the NYT API. I also applied it to NewsAPI for the same queries but date parameters limited to one month ago and no need for the additional pages loop as the NewsAPI allowed for more articles to be pulled at once.

By the end, I had four different word clouds that gave great initial insight into the discussion. The word clouds for “Virtual Reality” revealed major words related to prominent businesses in the industry (like Samsung, Apple, and Meta) and new applications of virtual reality (like gaming, smart glasses, and headsets). Meanwhile, the word clouds for “Cybersecurity” revealed an emphasis on words in the realm of geopolitics and modern data breaches; “China,” “Russia,” “state,” “government,” and “global” are just a few. Right away, it became clear why a boolean operator of “AND” between VR and Cybersecurity proved ineffective; these topics are not presently highly discussed in conjunction!

*TF-IDF*

Continuing the research, I proceeded to run Term Frequency - Inverse Document Frequency analysis to determine if certain words stood out across the articles. This began with converting the called articles into a PANDAS dataframe and employing BeautifulSoup to obtain the text. Then, using SK-learn and the TF-IDF vectorizer, I ran several codes to understand what terms were predominant in the articles and to see how similar or different the conversations around virtual reality and cybersecurity were. When I first ran a TF-IDF slice for words like “security” or “concern” on the topic of virtual reality, I received error messages because those words simply didn’t exist in the text of the all relevant articles. This simply confirmed my need to continue to research these topics individually to determine why the conversation was not merged between them.

Upon evaluating TF-IDF scores overall from the New York Times, terms like “impressed,” “tries,” “immersive,” and “investing” had the highest scores around 0.35 and 0.38. Generating the heatmap for the search term of “virtual reality” also reflected overall high scores, which indicate less common and more relatively unique terms. Conversely, for “cybersecurity,” the top TF-IDF terms were “computing,” “regulators,” “testify,” “executive” with a heatmap of overall low scores. Intriguingly, NewsAPI rendered similar results. “Virtual Reality” articles had top TF-IDF terms of “quest,” “world,” “glasses,” and “redefining” ranging from values of 0.38 - 0.48. Meanwhile, articles on cybersecurity generated high TF-IDF scores for words like “antivirus,” “fbi,” “finance,” and “cyberattack” but were much more heavily skewed towards low TF-IDF scores overall, which indicate more commonly utilized and discussed terms.

In short, the TF-IDF analysis revealed that news coverage on virtual reality is potentially more sparse and future-oriented, centering around current games and future developments, whereas news coverage on cybersecurity is much more prevalent and widespread but focuses on current policies, data breaches, and events rather than the future of the field.

*TOPIC MODELING*

Finally, I employed topic modeling to search for articles on either “Virtual Reality OR Cybersecurity.” As I was querying news APIs for articles and not a .txt file corpus of texts, I worked with AI to create code that would work in place of utilizing tomotopy. After deploying the code with NewsAPI, it scanned 100 articles and identified five topic lines that nearly all were related to VR applications like games, headsets, eyewear, and integration of VR into the world. This confirms that the current conversation around Virtual Reality is centered on products and newly developed and improved features. In the past month, this conversation even trumps the prevalence of discussions around cybersecurity; timeline bias necessitates noting that the search period does cover Black Friday. Even though I did put in a filter to remove results that included words like “Black Friday” or “Cyber Monday,” it is possible that the publishing skewed more towards topics of VR over Cybersecurity because of the tech products’ appeal to consumers.


### **Summary and Future Research**

Interestingly, the results from this research project answer the research questions differently than expected: Cybersecurity is not a current hot topic of conversation for virtual reality. The public is enamored with the applications and uses of virtual reality and while cybersecurity concerns are real for geopolitics and general financial cybercrime, it doesn’t appear to be highly discussed in the world of VR. Current news on virtual reality focuses on technological advancements and new consumer products while much of the current conversation around cybersecurity focuses on targeted hacks and data breaches happening around the world by cybercriminals.

```python
from nltk.stem import WordNetLemmatizer
from nltk.corpus import wordnet
from nltk import pos_tag
nltk.download('averaged_perceptron_tagger_eng')
nltk.download('punkt')
nltk.download('wordnet')
nltk.download('omw-1.4')

lemmatizer = WordNetLemmatizer()

def get_wordnet_pos(word):
    """Map POS tag to first character lemmatize() accepts"""
    tag = pos_tag([word])[0][1][0].upper()
    tag_dict = {
        'J': wordnet.ADJ,
        'N': wordnet.NOUN,
        'V': wordnet.VERB,
        'R': wordnet.ADV
    }
    return tag_dict.get(tag, wordnet.NOUN)
```

```python
lemmatized_string = ' '.join([lemmatizer.lemmatize(word, get_wordnet_pos(word)) for word in no_stops.split()])
```

```python
import altair as alt
import numpy as np

term_list = ['Security']

top_tfidf_plusRand = top_tfidf.copy()
top_tfidf_plusRand['tfidf'] = top_tfidf_plusRand['tfidf'] + np.random.rand(top_tfidf.shape[0])*0.0001

base = alt.Chart(top_tfidf_plusRand).encode(
    x = 'rank:O',
    y = 'document:N'
).transform_window(
    rank = "rank()",
    sort = [alt.SortField("tfidf", order="descending")],
    groupby = ["document"],
)

heatmap = base.mark_rect().encode(
    color = 'tfidf:Q'
)

circle = base.mark_circle(size=100).encode(
    color = alt.condition(
        alt.FieldOneOfPredicate(field='term', oneOf=term_list),
        alt.value('red'),
        alt.value('#FFFFFF00')
    )
)

text = base.mark_text(baseline='middle').encode(
    text = 'term:N',
    color = alt.condition(alt.datum.tfidf >= 0.23, alt.value('white'), alt.value('black'))
)

(heatmap + circle + text).properties(width = 600)
```

Topic Modeling

```python
import requests
import pandas as pd
import re
from sklearn.feature_extraction.text import CountVectorizer
from sklearn.decomposition import LatentDirichletAllocation

api_key = "4652ceb41ab94608bba2918a5f64bcb3"
url = "https://newsapi.org/v2/everything"

params = {
    "q": "Virtual Reality OR Cybersecurity",
    "from": "2024-11-18",
    "to": "2024-12-31",
    "pageSize": 100,
    "language": "en",
    "sortBy": "relevancy",
    "apiKey": api_key
}

response = requests.get(url, params=params)

if response.status_code == 200:
    data = response.json()
    articles = data.get("articles", [])
    df = pd.DataFrame([
        {
            "Title": article.get("title", ""),
            "Description": article.get("description", ""),
            "Content": article.get("content", "")
        }
        for article in articles
    ])
    print(f"Fetched {len(df)} articles")
else:
    print(f"Error: {response.status_code}, {response.text}")
    df = pd.DataFrame()

if not df.empty:
    df["Text"] = df["Title"] + " " + df["Description"] + " " + df["Content"]

    def preprocess_text(text):
        text = re.sub(r"http\S+|www\S+|https\S+", '', text, flags=re.MULTILINE)  # Remove URLs
        text = re.sub(r'\W', ' ', text)
        text = re.sub(r'\s+', ' ', text)
        text = re.sub(r'\d+', '', text)
        return text.lower().strip()

    df["Cleaned_Text"] = df["Text"].apply(lambda x: preprocess_text(x if x else ""))

    df = df[df["Cleaned_Text"].str.strip() != ""]
else:
    print("No articles fetched or available for processing.")

if not df.empty:
    vectorizer = CountVectorizer(max_df=0.95, min_df=2, stop_words="english")
    text_matrix = vectorizer.fit_transform(df["Cleaned_Text"])

    lda = LatentDirichletAllocation(n_components=5, random_state=42)
    lda.fit(text_matrix)

    def display_topics(model, feature_names, no_top_words):
        for topic_idx, topic in enumerate(model.components_):
            print(f"Topic {topic_idx + 1}:")
            print(" ".join([feature_names[i] for i in topic.argsort()[:-no_top_words - 1:-1]]))
            print()

    print("\n=== Topics Identified ===")
    no_top_words = 10
    display_topics(lda, vectorizer.get_feature_names_out(), no_top_words)

    topic_probs = lda.transform(text_matrix)
    df["Dominant_Topic"] = topic_probs.argmax(axis=1)
    df["Topic_Probability"] = topic_probs.max(axis=1)

    print("\n=== Articles with Assigned Topics ===")
    print(df[["Title", "Dominant_Topic", "Topic_Probability"]].head())
else:
    print("No valid data for topic modeling.")
```
